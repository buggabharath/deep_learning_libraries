```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
num_data = 1200
data = []
m = 0.2
c = 0.5
```

```python
for i in range(num_data):
    x = np.random.normal(0.0, 0.8)
    noise = np.random.normal(0.0, 0.4)
    y = m*x + c + noise
    data.append([x, y])
```

```python
# Spliting the date
x_data = [d[0] for d in data]
y_data = [d[1] for d in data]
```

```python
# plot the generated data
plt.plot(x_data, y_data, 'ro')
plt.title('Input data')
plt.show()
```

```python
# Generate weights and biases
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.zeros([1]))
```

```python
# Define equation for 'y'
y = W * x_data + b
```

```python
# Defing how to compute the loss
loss = tf.reduce_mean(tf.square(y - y_data))
```

```python
# Define the Gradient descent optiminzer
optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(loss)
```

```python
init = tf.initialize_all_variables()
```

```python
sess = tf.Session()
sess.run(init)
```

```python
num_iterations = 10
for step in range(num_iterations):
    sess.run(train)
```

```python
#print the progress

print('\n Iteration', step+1)
print('W = ', sess.run(W)[0])
print('b = ', sess.run(b)[0])
print('loss = ', sess.run(loss))
```

```python
plt.plot(x_data, y_data, 'ro')
plt.plot(x_data, sess.run(W) * x_data + sess.run(b))
```

```python
# set plotting parameter
plt.xlabel('Dimension 0')
plt.ylabel('Dimension 1')
plt.title('Iteration ' + str(step+1) + ' of ' + str(num_iterations))
plt.show()
```
